```python
import spacy
from spacy.matcher import Matcher
import logging

logging.basicConfig(level=logging.WARNING)  # Set the desired logging level


# Load the English language model
nlp = spacy.load("en_core_web_sm")

text = input("Hello, how can I help you? ")
doc = nlp(text)

# Print token text and POS tags
for token in doc:
    print(f"{token.text}  {token.pos_}")

matcher = Matcher(nlp.vocab)
fl_pattern = [{'LEMMA':'lose',"op":"?"}, {'LOWER': 'fat'},{"LOWER":"loss","OP":"?"},{"LOWER":"diet","OP":"*"}]
matcher.add("FAT_LOSS", [fl_pattern])

mg_pattern = [{'LEMMA':'gain',"op":"?"},{'LEMMA':'build',"op":"?"},{'LEMMA':'grow',"op":"?"}, {'LOWER': 'muscle'},{'LEMMA':'gain',"op":"?"},{'LEMMA':'build',"op":"?"},{'LEMMA':'grow',"op":"?"},{"LOWER":"diet","OP":"*"}]
matcher.add("MUSCLE_GAIN", [mg_pattern])

wg_pattern = [{'LEMMA':'gain',"op":"?"},{'LOWER': 'weight'},{'LEMMA':'gain',"op":"?"},{"LOWER":"diet","OP":"*"}]
matcher.add("WEIGHT_GAIN", [wg_pattern])

d_pattern = [{"LOWER":"help","OP":"?"},{"LOWER":"find","OP":"?"},{"LOWER":"diet"}]
matcher.add("DIET", [d_pattern])

md_pattern = [{"LOWER":"make","OP":"?"},{"LOWER":"help","OP":"?"},{"LOWER":"suggest","OP":"?"},{"LOWER":"mediterranean"},{"LOWER":"diet","OP":"!"}]
kd_pattern = [{"LOWER":"make","OP":"?"},{"LOWER":"help","OP":"?"},{"LOWER":"suggest","OP":"?"},{"LOWER":"keto"},{"LOWER":"diet","OP":"!"}]
pd_pattern = [{"LOWER":"make","OP":"?"},{"LOWER":"help","OP":"?"},{"LOWER":"suggest","OP":"?"},{"LOWER":"paleo"},{"LOWER":"diet","OP":"!"}]
vd_pattern = [{"LOWER":"make","OP":"?"},{"LOWER":"help","OP":"?"},{"LOWER":"suggest","OP":"?"},{"LOWER":"vegan"},{"LOWER":"diet","OP":"!"}]
dd_pattern = [{"LOWER":"make","OP":"?"},{"LOWER":"help","OP":"?"},{"LOWER":"suggest","OP":"?"},{"LOWER":"dash"},{"LOWER":"diet","OP":"!"}]
matcher.add("M_DIET", [md_pattern])
matcher.add("K_DIET", [kd_pattern])
matcher.add("P_DIET", [pd_pattern])
matcher.add("V_DIET", [vd_pattern])
matcher.add("D_DIET", [dd_pattern])
matches = matcher(doc)

for match_id, start, end in matches:
    matched_span = doc[start:end]
    print(matched_span.text, nlp.vocab.strings[match_id])
u_diet=nlp.vocab.strings[match_id]
```
